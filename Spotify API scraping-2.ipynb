```python
# Loading relevant packages
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
from datetime import datetime, timedelta
```

```python
# Log in to the website
driver = webdriver.Safari()
driver.get("https://accounts.spotify.com/en-GB/login?continue=https%3A%2F%2Fcharts.spotify.com/login")  

# Wait for the login page to load
time.sleep(1)

# Enter email
email_field = driver.find_element(By.ID, "login-username")
email_field.send_keys("INSERT_REAL_EMAIL")  

# Enter password
password_field = driver.find_element(By.ID, "login-password")
password_field.send_keys("INSERT_REAL_PASSWORD")  

# Click the log in button
login_button = driver.find_element(By.ID, "login-button")
login_button.click()

# Wait for the charts page to load
time.sleep(2)

# Get rid of cookies option
try:
    reject_button = WebDriverWait(driver, 5).until(
        EC.element_to_be_clickable((By.ID, "onetrust-reject-all-handler"))
    )
    reject_button.click()
    print("Cookies rejected.")
    WebDriverWait(driver, 5).until(
        EC.invisibility_of_element_located((By.ID, "onetrust-reject-all-handler"))
    )
except:
    print("No cookie pop-up found or already closed.")

# Select Ukraine from the global options
# Click the "Global" dropdown to activate it
search_box = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "input[placeholder='Global']"))
)
search_box.click()
time.sleep(1)

# Type "Ukraine" into the search box
search_box.send_keys("Ukraine")
time.sleep(1)  

# Click on "Ukraine" in the dropdown list
ukraine_option = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//div[contains(text(), 'Ukraine')]"))
)
ukraine_option.click()
print("Ukraine selected.")

# Wait for the page to reload
time.sleep(1)

# Select 'Weekly Top Tracks Ukraine'
# Wait for "Weekly Top Tracks Ukraine" box to appear and click it
weekly_top_songs = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//div[contains(@class, 'ChartsTile__Tile-sc-18xos3x-0') and contains(@style, 'rgb(119, 55, 149)')]"))
)
weekly_top_songs.click()

# Wait for the page to load
time.sleep(1)

# Open the calendar
date_picker = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "div.ChartsFilter__DatePickerFilter-vre17g-1"))
)
date_picker.click()
time.sleep(1)

# Click the left arrow until "January 2021" is reached
while True:
    # Check the calendar header
    calendar_header = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "strong"))
    )
    print(f"Current calendar header: {calendar_header.text}")  # Debugging output

    # Stop if the calendar header matches "January 2021"
    if calendar_header.text.strip() == "December 2020":
        print("Reached January 2021.")
        break

    # Click the left arrow to go to the previous month
    arrow_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CSS_SELECTOR, "svg.DayPickerNavigation_svg__horizontal"))
    )
    arrow_button.click()
    time.sleep(0.1)

# Click on January 1, 2021
specific_date = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//td[contains(@aria-label, 'Choose Friday, January 1, 2021 as your check-in date')]"))
)
specific_date.click()

# Wait for the page to update
time.sleep(1)

# Download the CSV file
button = WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "button[aria-labelledby='csv_download']"))
)
button.click()
```
